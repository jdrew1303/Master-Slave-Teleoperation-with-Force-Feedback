```python
from numpy import genfromtxt
import matplotlib.pyplot as plt
%matplotlib inline
from statistics import mean, median
plt.rcParams['figure.figsize'] = (20, 10)
import matplotlib.patches as mpatches
import numpy as np
import math
```

```python
data = genfromtxt('P5D0.csv', delimiter=';')
e = data[:,1:7]

fig, ax1 = plt.subplots()
ax1.set_xlabel('Time [sec]', size=30)
ax1.set_ylabel('Desired Position [rad]', size=30)

t = np.linspace(0,len(e)/100,len(e))
plt.plot(t,e[:,3],color='#1f77b4',linewidth=3)
plt.plot()

ax1.set_xlim([min(t[:]),max(t)])
ax1.set_ylim([-1,1.4])

ax1.axvspan(63, 74.5, color='hotpink', alpha=0.4)
ax1.axvspan(92, 101, color='hotpink', alpha=0.4)

ax1.tick_params(axis = 'both', which = 'major', labelsize = 24)
plt.minorticks_on()
plt.grid()
plt.savefig('/home/robotronics/Pictures/direction_change.png', format='png', dpi=100)
```
